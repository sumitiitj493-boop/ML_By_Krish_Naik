#### 1.1. Overfitting Problem

```
Scenario: 
   When you have only 2 data points, the best fit line passes through both
   points perfectly

Problem:
   When new test data is added, error increases significantly!

### Biasâ€“Variance Tradeoff

| Data Type      | Accuracy              | Interpretation |
|---------------|-----------------------|----------------|
| **Training Data** | Very High (99â€“100%) | Low Bias |
| **Test Data**     | Low                 | High Variance |

> âš ï¸ **Important:**  
> If you get **100% training accuracy**, definitely consider that the model is **overfitting**!

> Never aim for perfect training accuracy


#### 1.2 The Solution : Ridge Regression

#### Defining Cost Function

```
Ridge regression adds a penalty term to the linear regression cost function to reduce overfitting.
It shrinks coefficients towards zero but never makes them exactly zero.

This actually preventing the best-fit line from passing exactly through all the training point.



> When you have only 2 data points, the best fit line passes through both points perfectly.When new test data is added, error increases significantly!

```
 But..But..But Wait
 Why we fear when we have Ridge Regression?

    Even in this case the Ridge Regression will not overfit the data.Instead it finds the best fit line that minimizes the cost 
    function.

<div style="border:2px solid #247026ff; padding:10px; border-radius:8px;"><b>Always Remember :</b>

â€¢ When best fit line passes through all points perfectly â‡’ First term becomes zero

â€¢ This means model is overfitted

â€¢ By adding Î»Â·Î¸2, we ensure cost function is never zero
</div>

#### 1.3 Relationship Between Î» and Slope

<div style="border:2px solid #247026ff; padding:10px; border-radius:8px;"><b>Always Remember :</b>
</div>

```

    1. Training Data: For training model + hyperparameter tuning
    2. Test Data: NEVER shown to model during training. Only used to check final performance.
    3. Validation Data: Created from training data for hyperparameter tuning

##### 4.1 CV Type 1 : Leave One Out Cross-Validation (LOOCV)


<mark style="background-color: #367024ff; color: #fff; padding: 5px 10px; border-radius: 5px;">Method: Take one record as validation, remaining as training
</mark>
```

If training data has 500 records:
â€¢ Experiment 1: Record 1 = Validation, Records 2-500 = Training
â€¢ Experiment 2: Record 2 = Validation, Records 1,3-500 = Training
â€¢ ...
â€¢ Experiment 500: Record 500 = Validation, Records 1-499 = Training


<mark style="background-color: #367024ff; color: #fff; padding: 5px 10px; border-radius: 5px;">Total experiments = Total records</mark>

<details><summary><b> Disadvantages of LOOCV</b></summary>

1. High Complexity: 
   
        If 5000 records â‡’ 5000 experiments needed 
   
2. Leads to Overfitting:

        â€¢ Training data is huge â‡’ High training accuracy
        â€¢ Validation data is only 1 record â‡’ Poor validation
        â€¢ Model wonâ€™t perform well on new test data
</details>

##### 4.2 CV Type 2 : Leave P Out Cross-Validation

<mark style="background-color: #367024ff; color: #fff; padding: 5px 10px; border-radius: 5px;">
 Method: Instead of one, set P records as validation
P value can be: 10, 20, 30 (selected as hyperparameter)
</mark>

```
Same process as LOOCV, just with P records instead of 1

##### 4.3 CV Type 3 : K-Fold Cross-Validation

<mark style="background-color: #367024ff; color: #fff; padding: 5px 10px; border-radius: 5px;">
 Method: Split data into K equal parts(folds)
</mark>

``` 
  Formula : Fold Size = Total Records / K value.
  Example :
    Total Records = 500
    K value = 5
    Fold Size = 500 / 5 = 100
    Each fold will have 100 records.



<details><summary><b> K-Fold Results</b></summary>
After K experiments :  

    â€¢ Calculate average of all accuracies
    â€¢ Find maximum and minimum accuracy
    â€¢ Report: "Average accuracy is X%"
</details>

<mark style="background-color: #367024ff; color: #fff; padding: 5px 10px; border-radius: 5px;"><b>Note:</b> </mark>

```
 Number of experiments = K value

##### 4.4  Type 4: Stratified K-Fold Cross-Validation

<mark style="background-color: #367024ff; color: #fff; padding: 5px 10px; border-radius: 5px;"><b>Problem with K-Fold:</b> </mark>

<details><summary><b>For Classification Problems:</b></summary>

    â€¢ When selecting validation data sequentially, only ONE type of category may appear
    â€¢ Example: In binary classification, validation set might have all 1s or all 0s
    â€¢ Model wonâ€™t train properly!
</details>

<mark style="background-color: #367024ff; color: #fff; padding: 5px 10px; border-radius: 5px;"><b>Stratified K-Fold Solution :</b> </mark> 

<details><summary><b>Stratified K-Fold ensures:</b></summary>

    â€¢ Output labels are evenly distributed in each validation fold
    â€¢ Example: If original data has 60% class-1 and 40% class-0
    â€¢ Each validation fold will have approximately 60-40 ratio
    â€¢ Maintains balanced data in validation set
</details>
   

<mark style="background-color: #367024ff; color: #fff; padding: 5px 10px; border-radius: 5px;"><b>Remember:</b>
 </mark>

 ```
 Use Stratified K-Fold for classification problems to ensure balanced class distribution in each fold.

##### 4.5 CV Type 5 : Time Series Cross-Validation

<mark style="background-color: #367024ff; color: #fff; padding: 5px 10px; border-radius: 5px;"><b>Use Case :</b> 
</mark>  
>Data that changes over time (e.g., product reviews, stock prices)

```
