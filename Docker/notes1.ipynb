### ğŸ“¦ What is a Container?

<div style="background: #e3f2fd; border-left: 5px solid #0db7ed; padding: 20px; border-radius: 0 10px 10px 0; margin: 20px 0;">
    <h3 style="color: #0277bd; margin-top: 0;">ğŸ“˜ Container Definition</h3>
    <p style="color: #333; font-size: 1.1em;"><strong>Containers</strong> are a way to <strong>package applications</strong> with all the necessary <strong>dependencies</strong> and <strong>configuration</strong>.</p>
    <h4 style="color: #0277bd;">Key Properties of Containers:</h4>
    <ol style="color: #333;">
        <li><strong>Portable Artifact:</strong> Can be easily shared and moved to any environment</li>
        <li><strong>Self-Contained:</strong> Includes all dependencies within the package</li>
        <li><strong>Consistent:</strong> Works the same way in Dev, QA, and Production</li>
    </ol>
</div>

<div style="background: #e8f5e9; border-left: 5px solid #4caf50; padding: 20px; border-radius: 0 10px 10px 0; margin: 20px 0;">

        <li>Properly <strong>in sync</strong> across all environments</li>
        <li>All developers work with <strong>same configurations</strong></li>
    </ul>
</div>

#### 1. What is 0.0.0.0?

#### 2. What is localhost?
 - The actual IP of localhost is 127.0.0.1.
 - It always points to your own machine.

#### 3. Difference Between 0.0.0.0 and 127.0.0.1?
 - 0.0.0.0 â†’ Listen on all interfaces (accessible from outside if firewall allows).
 - 127.0.0.1 â†’ Access locally from browser (only your machine).
 - Both effectively work for local testing.

#### 4. Meaning of 80->80?
 - Format: <host-port> -> <container-port>
 - Left side (Host Port = 80): Port on your local machine (browser sends request here).
 - Right side (Container Port = 80): Port inside the Docker container where the app is running.

 Open your browser and go to:
 - http://localhost:80

### ğŸ“„ What is a Dockerfile?

<div style="background: #e3f2fd; border-left: 5px solid #0db7ed; padding: 20px; border-radius: 0 10px 10px 0; margin: 20px 0;">
    <h3 style="color: #0277bd; margin-top: 0;">ğŸ“˜ Dockerfile</h3>
    <p style="color: #333;">A <strong>Dockerfile</strong> is a text file containing instructions to build a Docker image. It defines:</p>
    <ul style="color: #333;">
        <li>Base image to use</li>
        <li>Files to copy into the image</li>
        <li>Dependencies to install</li>
        <li>Commands to run</li>
        <li>Port to expose</li>
        <li>Startup command</li>
    </ul>
</div>

### ğŸ“ Dockerfile Commands

| Command | Description |
|---------|-------------|
| `FROM` | Specifies the base image (e.g., Linux, Python) |
| `WORKDIR` | Sets the working directory inside the container |
| `COPY` | Copies files from local machine to container |
| `RUN` | Executes commands during image build |
| `EXPOSE` | Declares the port the container will use |
| `ENV` | Sets environment variables |
| `CMD` | Specifies the command to run when container starts |
