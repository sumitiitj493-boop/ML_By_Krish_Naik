## 2.2 What is a Container?

<div style="background: linear-gradient(to right, #dbeafe, #eff6ff); padding: 25px; border-radius: 12px; border-left: 5px solid #0969da; margin: 20px 0;">
    <h3 style="color: #0969da; margin-top: 0;">ğŸ“˜ Container Definition</h3>
    <p style="font-size: 1.1em; color: #1e3a5f;"><strong>Containers</strong> are a way to <strong>package applications</strong> with all necessary <strong>dependencies</strong> and <strong>configuration</strong>.</p>
    <h4 style="color: #0969da;">Key Properties:</h4>
    <ol style="color: #1e3a5f;">
        <li><strong>Portable Artifact:</strong> Can be easily shared and moved to any environment</li>
        <li><strong>Self-Contained:</strong> Includes all dependencies within the package</li>
        <li><strong>Consistent:</strong> Works the same way in Dev, QA, and Production</li>
    </ol>
</div>

| `EXPOSE` | Declares the port the container will use |

| `CMD` | Specifies the command to run when container starts |

## 7.4 Creating the Dockerfile

````python
# Dockerfile

# Base image - Python 3.8 on Alpine Linux (small size)
FROM python:3.8-alpine

# Copy all files from current directory to /app in container
COPY . /app

# Set working directory
WORKDIR /app

# Install dependencies
RUN pip install -r requirements.txt

# Command to run the application
CMD ["python", "app.py"]
````

## 9.3 Complete Docker Compose Example

````python
# Flask App with Redis - app.py

from flask import Flask
import redis

app = Flask(__name__)
cache = redis.Redis(host='redis', port=6379)

def get_hit_count():
    return cache.incr('hits')

@app.route('/')
def hello():
    count = get_hit_count()
    return f'Hello! I have been seen {count} times.'

if __name__ == '__main__':
    app.run(host='0.0.0.0', port=5000)
````

````python
# Dockerfile with Environment Variables

FROM python:3.7-alpine

WORKDIR /code

# Environment variables for Flask
ENV FLASK_APP=app.py
ENV FLASK_RUN_HOST=0.0.0.0

COPY . .

RUN pip install -r requirements.txt

EXPOSE 5000

CMD ["flask", "run"]
````

````python
# docker-compose.yml

version: "3.0"

services:
  web:
    build: .
    image: web-app
    ports:
      - "8000:5000"
  
  redis:
    image: redis
````
