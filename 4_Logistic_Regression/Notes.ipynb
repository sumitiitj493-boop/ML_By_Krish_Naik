#### 7.3 Imbalanced Dataset Problem

>##### - **When NOT to Use Accuracy ?**
```
The Answer is in case of imbalanced dataset.
Example: Consider we have a dataset with 1000 data points:
- 900 data points belong to class 1
- 100 data points belong to class 0
  
  In this case Ratio of class 1 to class 0 is 9:1
  So, if we always predict class 1, we will get 90% accuracy.
  But, this is not a good model.
  Because, we are not able to identify class 0.

And one more thing is we don't even know in 900 sample input which one gives TP(1 is output so input will also be 1) and which one gives FP(1 is output so input will be 0).SO definetly we will have to reduce the FP(1 is output so input will be 0) as much as possible.
```

#### 7.6 F-Beta Score

>##### **F-beta Score Formula**

$$
        F_{\beta} = \frac{(1+\beta^2)\cdot \text{Precision} \cdot \text{Recall}}
        {\beta^2 \cdot \text{Precision} + \text{Recall}}
$$

>- ##### Choosing Beta (Î²) Value

| Condition                          | Î² Value | Score Name |
|-----------------------------------|--------:|-----------|
| FP & FN equally important         | 1       | F1 Score  |
| FP more important than FN         | 0.5     | F0.5 Score|
| FN more important than FP         | 2       | F2 Score  |


>- #####  F1 Score (Harmonic Mean)

- When Î² = 1, F1 Score is the harmonic mean of Precision and Recall.
  $$
  F1 = \frac{2 \cdot Precision \cdot Recall}{Precision + Recall}
  $$

>- ##### F0.5 Score 

- When FP is more important than FN, choose Î² = 0.5.
  $$
  F0.5 = \frac{1.25 \cdot Precision \cdot Recall}{0.25 \cdot Precision + Recall}
  $$

>- ##### F2 Score

- When FN is more important than FP, choose Î² = 2.
  $$
  F2 = \frac{5 \cdot Precision \cdot Recall}{4 \cdot Precision + Recall}
  $$

>### 8.  Multi-Class Classification: One vs Rest (OvR)

```
Problem Statement
What if we have more than 2 output categories?
```
<details><summary><b>Solution :</b></summary>
Use One vs Rest (OvR) approach with Logistic Regression

- ##### One vs Rest (OvR) approach
  ```
    â€¢ For n output categories, create n binary classification models:
    â€¢ Each model differentiates one class vs all other classes combined
  ```
  <mark style="background-color: #367024ff; color: #fff; padding: 5px 10px; border-radius: 5px;">Important:
   3 categories â‡’ 3 models</mark>

  ```
  Model M1 : A vs (B+C)
  Model M2 : B vs (A+C)
  Model M3 : C vs (A+B)
  ```



| Features       | Output | O1 | O2 | O3 |
|----------------|--------|----|----|----|
| f1, f2, f3     | O1     | 1  | 0  | 0  |
| f1, f2, f3     | O2     | 0  | 1  | 0  |
| f1, f2, f3     | O3     | 0  | 0  | 1  |
| f1, f2, f3     | O1     | 1  | 0  | 0  |
| f1, f2, f3     | O3     | 0  | 0  | 1  |
| f1, f2, f3     | O2     | 0  | 1  | 0  |


---

Here, since we are dealing with a **multi-class classification problem**, the output can belong to **three different classes**, namely **O1, O2, and O3**.

For each data point, the **input features (f1, f2, f3)** remain the same in structure, but the **output class changes** depending on the sample.

To represent these multiple classes numerically, we use **one-hot encoding**:
- If the output class is **O1**, it is represented as **[1, 0, 0]**
- If the output class is **O2**, it is represented as **[0, 1, 0]**
- If the output class is **O3**, it is represented as **[0, 0, 1]**

This representation ensures that:
- Each class is treated independently
- Only one class is active (1) at a time
- The model can clearly distinguish between different output classes


---

#### Multi-Class Classification using One-vs-Rest (OvR)

>>##### Problem Setup
- This is a **multi-class classification problem**
- Output classes:
  - **O1**
  - **O2**
  - **O3**
- Input features for every data point:
  - **f1, f2, f3**

Since **logistic regression is a binary classifier**, it cannot directly handle more than two classes.

So, we convert the multi-class problem into **multiple binary classification problems** using the **One-vs-Rest (OvR)** approach.

---

>>##### One-Hot Encoding of Output
Each output class is represented as a binary vector:

- **O1 â†’ [1, 0, 0]**
- **O2 â†’ [0, 1, 0]**
- **O3 â†’ [0, 0, 1]**

This ensures that:
- Only one class is active at a time
- Each class is treated independently

---

>>##### Individual Models Explanation

##### Model M1 (O1 vs Rest)
- **Input:** f1, f2, f3
- **Target Output:** O1 column

This model answers:
> Is the given input **O1 or NOT O1**?

- O1 â†’ 1  
- Not O1 â†’ 0  

---

##### Model M2 (O2 vs Rest)
- **Input:** f1, f2, f3
- **Target Output:** O2 column

This model answers:
> Is the given input **O2 or NOT O2**?

- O2 â†’ 1  
- Not O2 â†’ 0  

---

##### Model M3 (O3 vs Rest)
- **Input:** f1, f2, f3
- **Target Output:** O3 column

This model answers:
> Is the given input **O3 or NOT O3**?

- O3 â†’ 1  
- Not O3 â†’ 0  

---

>>##### How Final Prediction is Made

For a **new input (f1, f2, f3)**:
- M1 predicts probability of **O1**
- M2 predicts probability of **O2**
- M3 predicts probability of **O3**

The **final class** is the one with the **highest predicted probability**.

Example: 
M1 = 0.25,
M2 = 0.20,
M3 = 0.55

Final Output = **O3**

---


<details>
<summary><b>Prediction Process : </b></summary> 

     1. Pass new test data to all models
     2. Each model outputs a probability
     3. Select the model with highest probability
     4. Output class = class that model was trained for
</details>
