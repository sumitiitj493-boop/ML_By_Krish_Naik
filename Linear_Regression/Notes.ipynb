>### 1. Error Metrices : MSE, MAE, RMSE

#### Why do we need Error Metrices?

>1. To evaluate the performance of a linear regression model.
   
>2. To compare the performance of different linear regression models.

#### 1.1 Mean Squared Error (MSE)

#### Advantages & Disadvantages of MSE

<details>
<summary><b>Click to expand: MSE Advantages</b></summary>

1. Differentiable at all points- Can calculate slope at every point
2. One local and one global minima- Convex function, no multiple minimas
3. Converges faster- No local minima to get stuck

</details>


<details><summary><b>Click to expand: MSE Disadvantages</b></summary>

1. Not Robust to Outliers : 
   
   Squaring large errors penalizes outliers heavily, causing the best-fit line to shift dramatically.

1. Unit Problem :
    
   Example:
   If the target value y is measured in lakhs, then MSE is measured in lakhsÂ².

</details>

#### 1.2 Mean Absolute Error (MAE)

#### Advantages & Disadvantages of MAE

<details><summary><b>Click to expand: Advantages</b></summary>

1. Robust to Outliers : 
 
   >MAE is less sensitive to outliers than MSE because it does not square the errors.

2. Same Units as Output :
   
    >MAE is more interpretable than MSE because it is in the sameunit as the target variable.
</details>

<details><summary><b>Click to expand: Disadvantages of MAE</b></summary>

1. Not Differentiable at Zero :
    >MAE is not differentiable at 0, which can make optimization difficult.

2. Slower Convergence :
    >MAE is slower to converge than MSE because it does not have the smoothness of MSE.
</details>

>#### <mark>Solution for Not Differentiable at Zero</mark>

    Use subgradient descent to approximate the derivative : 

    Subgradient descent is a variant of gradient descent that handles non-differentiable functions by using a subgradient instead of the true gradient.

    In simple words, subgradient descent uses a step that is not exactly the gradient but a direction that points towards the minimum.

```text
â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
â”‚               FULL DATASET                 â”‚
â”‚               (1000 records)               â”‚
â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜
                     â”‚
          â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”´â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
          â”‚                     â”‚
          â–¼                     â–¼
â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”   â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
â”‚   TRAINING DATA   â”‚   â”‚     TEST DATA      â”‚
â”‚     (70%)         â”‚   â”‚      (30%)         â”‚
â”‚   700 records     â”‚   â”‚   300 records      â”‚
â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜   â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜
          â”‚                     â”‚
          â”‚                     â”‚
          â”‚                     â””â”€â”€â–º Final Evaluation
          â”‚                          (Never used in training)
          â”‚
          â–¼
â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
â”‚   TRAIN + VALID   â”‚
â”‚   SPLIT           â”‚
â”‚   (700 records)   â”‚
â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜
          â”‚
    â”Œâ”€â”€â”€â”€â”€â”´â”€â”€â”€â”€â”€â”
    â”‚           â”‚
    â–¼           â–¼
â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”  â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
â”‚ TRAIN SET    â”‚  â”‚ VALIDATION SET â”‚
â”‚ 560 records  â”‚  â”‚ 140 records    â”‚
â”‚              â”‚  â”‚                â”‚
â”‚ Model        â”‚  â”‚ Hyperparameter â”‚
â”‚ Training     â”‚  â”‚ Tuning         â”‚
â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜  â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜




#### 2.1 Dataset Splitting Strategy

##### Critical Rule: Data Leakage Prevention

- Test data is only for final checking
- It should not be used while building the model
- Using test data early gives false good results
- Keeping it separate shows true model performance


---

#### 2.2 Overfitting vs Underfitting

>>##### Scenario 1: Generalized Model (Goal)

### Ideal Performance

| Dataset   | Accuracy |
|-----------|----------|
| Training  | 90%      |
| Test      | 85%      |

**Status:** Good â€” training and test accuracies are high and close to each other.  
**Terminology:** Low Bias, Low Variance


>>##### Scenario 2: Overfitting (Problem)

### Overfitting Signs

| Dataset   | Accuracy        |
|-----------|-----------------|
| Training  | 90% (High)      |
| Test      | 50% (Low)       |

**Problem:** The model learns the training data too well but performs poorly on new data.  
**Terminology:** Low Bias, High Variance


>>##### Scenario 3: Underfitting (Problem)

### Underfitting Signs

| Dataset   | Accuracy        |
|-----------|-----------------|
| Training  | 50% (Low)       |
| Test      | 45% (Low)       |

**Problem:** The model is too simple and cannot learn important patterns from the data.  
**Terminology:** High Bias, High Variance


<details><summary>Key Takeaway:</summary>

   >Goal: Always aim for Low Bias, Low Variance (generalized model).

   >Use R, Adjusted R, MSE, MAE, RMSE together to diagnose issues.


</details>


>### 3. Ordinary Least Squares (OLS)- Mathematical Derivation

####  Introduction to OLS

##### What is OLS? Why do we use it?

      OLS is a method to estimate the coefficients of a linear regression model.

      We use OLS because it is the most common method to estimate the coefficients of a linear regression model.

####  Step 1: Define the Error Function

#### Step 2: Partial Derivatives

#### Step 3: Centered form (MOST IMPORTANT)

#### Final Step:  OLS Formula Summary

>### 4 Practical Implementation in Python
