# 2. Bagging (Bootstrap Aggregating)

---

<div style="background-color: #DBEAFE; border-left: 5px solid #2563EB; padding: 20px; border-radius: 0 10px 10px 0; margin: 15px 0;">
    <h3 style="color: #1D4ED8; margin-top: 0;">ðŸ“¦ Understanding Bagging</h3>
    <ol style="color: #1F2937; font-size: 1.05em; line-height: 1.8;">
        <li>Create multiple <strong>base learners</strong> (ML models)</li>
        <li>Give each model a <strong>different sample</strong> of the dataset</li>
        <li>Train all models <strong>in parallel</strong></li>
        <li>Combine predictions using <strong>voting/averaging</strong></li>
    </ol>
</div>

### ðŸ—ï¸ Bagging Architecture




<div style="background-color: #FEE2E2; border: 2px solid #DC2626; padding: 15px; border-radius: 10px; margin: 15px 0;">
    <h4 style="color: #991B1B; margin-top: 0;">âš ï¸ Important Point</h4>
    <p style="color: #7F1D1D; font-size: 1.1em;">In Bagging, all base learners are trained <strong>PARALLEL</strong> (simultaneously), not sequentially!</p>
</div>

### ðŸ“ Formulas

**Classification (Majority Voting):**
$$\hat{y} = \text{mode}\{h_1(x), h_2(x), \ldots, h_n(x)\}$$

**Regression (Average):**
$$\hat{y} = \frac{1}{n}\sum_{i=1}^{n} h_i(x)$$

Where $h_i(x)$ is the prediction of the $i^{th}$ base learner.

# 3. Boosting (Sequential Learning)

---

<div style="background-color: #F3E8FF; border-left: 5px solid #7C3AED; padding: 20px; border-radius: 0 10px 10px 0; margin: 15px 0;">
    <h3 style="color: #6D28D9; margin-top: 0;">ðŸš€ Understanding Boosting</h3>
    <ol style="color: #1F2937; font-size: 1.05em; line-height: 1.8;">
        <li>Multiple <strong>weak learners</strong> are combined</li>
        <li>Models are trained <strong>sequentially</strong></li>
        <li>Each model <strong>learns from mistakes</strong> of previous models</li>
        <li>Combined to form a <strong>strong learner</strong></li>
    </ol>
</div>

### ðŸ”— Boosting Architecture



<div style="background-color: #ECFDF5; border: 2px solid #059669; padding: 15px; border-radius: 10px; margin: 15px 0;">
    <h4 style="color: #047857; margin-top: 0;">ðŸŽ¯ The Magic of Boosting</h4>
    <ul style="color: #064E3B; line-height: 1.8;">
        <li>First model trains on <strong>entire dataset</strong></li>
        <li><strong>Incorrectly predicted</strong> records are passed to next model</li>
        <li>Each subsequent model <strong>focuses on hard cases</strong></li>
        <li>Final prediction combines all models' <strong>weighted strengths</strong></li>
    </ul>
</div>

### ðŸŽ“ How Does Random Forest Reduce Variance?

<div style="background-color: #DBEAFE; padding: 20px; border-radius: 10px; margin: 15px 0;">

**Result:**
$$\text{High Variance} \xrightarrow{\text{Multiple Trees}} \textbf{Low Variance}$$
