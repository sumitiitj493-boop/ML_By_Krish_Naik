In unsupervised machine learning, we don't have a "label" (output feature) to predict. We just have a set of features (input features) to use to train our model.

>### 5 Instance Based Learning vs Model Based Learning

<mark>Core Concept</mark>

When solving a use case (regression/classification), the ML model learns about
data in two ways:

>1. Instance Based Learning (Memorizing)
>2. Model Based Learning (Generalizing)

This is based on the learning patterns of the model.

#### 5.1 Instance Based Learning

<details><summary>Instance Based Learning</summary>

â€¢ Model is completely dependent on training data

â€¢ For every prediction, it depends on the specific data

â€¢ Just like a domain expert- memorizing the data

â€¢ Model does NOT understand patterns within data

â€¢ Looks at surrounding data points for prediction
</details>

Example Algorithm: K-Nearest Neighbor (KNN)

#### 5.2 Model Based Learning

<details><summary>Model Based Learning</summary>

â€¢ Tries to understand the pattern within the data

â€¢ Creates a generalization method for predictions

â€¢ Creates decision boundaries

â€¢ Works well for new, unseen data

â€¢ Far better than instance based techniques
</details>

Example Algorithm: Logistic Regression

<b>NOTE:</b> 
Generalizing is better than Memorizing!

But for some use cases, memorizing techniques can be used.
</div>


#### 5.3 Comparing Instance Based and Model Based Learning

#### <mark> Why Model Based is Faster?</mark>

>Model Based:

â€¢ Model is in serialized format

â€¢ Mathematical equations get triggered immediately

â€¢ Output is quick

>Instance Based:

â€¢ Must look at surrounding data points

â€¢ Uses distance parameters (e.g., Euclidean distance)

â€¢ Then makes decision â†’ Slower!
